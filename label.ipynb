```python
import os
import shutil
import yaml

# Directories for Dataset 01
raw_image_dir = r"D:\Wildlife dataset\data_01"
new_image_dir = r"D:\Wildlife dataset\wildlife_merged_dataset\data_01"

# Load class names from YAML
yaml_path = os.path.join(raw_image_dir, 'data.yaml')
with open(yaml_path, 'r') as file:
    data_yaml = yaml.safe_load(file)
    dataset_classes = data_yaml['names']  # List of class names (index 0 = 'animal', index 1 = 'bear', etc.)

# Define subfolders
splits = ['train', 'valid', 'test']
dataset_name = "01_dataset"  # Dataset name to include in the filenames

def rename_and_copy_images_and_labels(image_dir, label_dir, new_image_dir, new_label_dir, class_names, dataset_name):
    # Ensure folders exist
    os.makedirs(new_image_dir, exist_ok=True)
    os.makedirs(new_label_dir, exist_ok=True)

    image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]
    label_files = [f for f in os.listdir(label_dir) if f.endswith('.txt')]

    image_files.sort()
    label_files.sort()

    class_counters = {i: 1 for i in range(len(class_names))}

    for image_file, label_file in zip(image_files, label_files):
        label_path = os.path.join(label_dir, label_file)

        with open(label_path, 'r') as f:
            lines = f.readlines()
            if not lines:
                print(f"Empty label file: {label_file}")
                continue
            label_class_index = int(lines[0].split()[0])

        class_name = class_names[label_class_index]
        count = class_counters[label_class_index]

        # Construct new base name with dataset name
        new_base_name = f"{label_class_index}id_{count:03d}_{class_name}_{dataset_name}"
        new_image_name = new_base_name + os.path.splitext(image_file)[1]
        new_label_name = new_base_name + ".txt"

        shutil.copy(os.path.join(image_dir, image_file), os.path.join(new_image_dir, new_image_name))
        shutil.copy(label_path, os.path.join(new_label_dir, new_label_name))

        class_counters[label_class_index] += 1

        print(f"{image_file} -> {new_image_name}")
        print(f"{label_file} -> {new_label_name}")

# Run the function for all three splits
for split in splits:
    raw_images_dir = os.path.join(raw_image_dir, split, 'images')
    raw_labels_dir = os.path.join(raw_image_dir, split, 'labels')

    new_images_dir = os.path.join(new_image_dir, split, 'images')
    new_labels_dir = os.path.join(new_image_dir, split, 'labels')

    rename_and_copy_images_and_labels(raw_images_dir, raw_labels_dir, new_images_dir, new_labels_dir, dataset_classes, dataset_name)
```

```python
import os
import shutil

# Define the paths of the datasets
dataset_paths = [
    r"D:\Wildlife dataset\wildlife_merged_dataset\data_01",
    r"D:\Wildlife dataset\wildlife_merged_dataset\data_02",
    r"D:\Wildlife dataset\wildlife_merged_dataset\data_03",
    r"D:\Wildlife dataset\wildlife_merged_dataset\data_04"
]

# Define the output directory for the merged dataset
output_dir = r"D:\Wildlife dataset\merged_dataset"

# Define subdirectories to merge
subdirs = ['train', 'test', 'valid']

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Function to copy the dataset files to the merged directory
def copy_files(dataset_path, subdir):
    # Define paths for images and labels
    image_dir = os.path.join(dataset_path, subdir, 'images')
    label_dir = os.path.join(dataset_path, subdir, 'labels')

    # Destination directories for merged images and labels
    dest_image_dir = os.path.join(output_dir, subdir, 'images')
    dest_label_dir = os.path.join(output_dir, subdir, 'labels')

    # Create subdirectories if they don't exist
    os.makedirs(dest_image_dir, exist_ok=True)
    os.makedirs(dest_label_dir, exist_ok=True)

    # Copy images and labels from the dataset to the merged directory
    for filename in os.listdir(image_dir):
        # Copy image file
        image_file = os.path.join(image_dir, filename)
        if os.path.isfile(image_file):
            shutil.copy(image_file, dest_image_dir)

        # Copy corresponding label file
        label_file = os.path.join(label_dir, filename.replace('.jpg', '.txt').replace('.jpeg', '.txt').replace('.png', '.txt'))
        if os.path.isfile(label_file):
            shutil.copy(label_file, dest_label_dir)

# Loop through datasets and merge them
for dataset_path in dataset_paths:
    for subdir in subdirs:
        copy_files(dataset_path, subdir)

print("Merging complete!")
```

```python
import os
import yaml

# Load the data.yaml file to extract class names and indices
data_yaml = """
names:
  0: animal
  1: bear
  2: bull
  3: camel
  4: cheetah
  5: crocodile
  6: deer
  7: dog
  8: elephant
  9: fox
  10: giraffe
  11: hippo
  12: lion
  13: monkey
  14: pig
  15: poacher
  16: ranger
  17: rhino
  18: tiger
  19: turkey
  20: weapon
  21: zebra
  22: horse
  23: hyena
  24: sheep
  25: cow
  26: sloth bear
  27: wild-boar
  28: leopard
  29: hedgehog
  30: lynx
  31: ostrich
  32: human
nc: 33
test: /path/to/test/images
train: /path/to/train/images
val: /path/to/val/images
"""

# Parse the YAML content
yaml_data = yaml.safe_load(data_yaml)
valid_class_indices = set(yaml_data['names'].keys())  # Set of valid class indices (0, 1, 2, ..., 32)

# Directory paths for labels
label_dirs = [
    r"D:\Wildlife dataset\merged_dataset\train\labels",
    r"D:\Wildlife dataset\merged_dataset\test\labels",
    r"D:\Wildlife dataset\merged_dataset\valid\labels"
]

# Function to check if the class IDs in label files match those in data.yaml
def check_label_files(label_dir):
    mismatched_files = []
    for label_file in os.listdir(label_dir):
        if label_file.endswith('.txt'):
            with open(os.path.join(label_dir, label_file), 'r') as file:
                lines = file.readlines()
                for line in lines:
                    class_id = int(line.split()[0])  # Get the class ID from the first element
                    if class_id not in valid_class_indices:
                        mismatched_files.append((label_file, class_id))
    return mismatched_files

# Check all label directories
for label_dir in label_dirs:
    mismatched_files = check_label_files(label_dir)
    if mismatched_files:
        print(f"Found mismatched class IDs in {label_dir}:")
        for file, class_id in mismatched_files:
            print(f"File: {file}, Mismatched Class ID: {class_id}")
    else:
        print(f"All class IDs are valid in {label_dir}.")
```
